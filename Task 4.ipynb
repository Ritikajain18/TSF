```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
x = iris.data
y = iris.target
```

```python
clf = tree.DecisionTreeClassifier()
model = clf.fit(x,y)
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(model,filled = True)
```
